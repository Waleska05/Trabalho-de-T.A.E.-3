```r
install.packages("fastDummies")
install.packages("dplyr")
install.packages("caret")
install.packages("ggplot2")
install.packages("pROC")
install.packages("rpart")
install.packages("rpart.plot")
install.packages("randomForest")

library(fastDummies)
library(dplyr)
library(caret)
library(ggplot2)
library(pROC)
library(rpart)
library(rpart.plot)
library(randomForest)
```

```r
dados <- read.csv("heart.csv")

dados_ohe <- fastDummies::dummy_cols(
  dados,
  select_columns = c("Sex","ChestPainType","RestingECG","ExerciseAngina","ST_Slope"),
  remove_first_dummy = TRUE,
  remove_selected_columns = TRUE
)

num_cols <- c("Age","RestingBP","Cholesterol","FastingBS","MaxHR","Oldpeak")
dados_ohe[num_cols] <- scale(dados_ohe[num_cols])

dados$HeartDisease <- as.factor(dados$HeartDisease)
```

```r
modelo_linear <- lm(
  Cholesterol ~ .,
  data = dados_ohe
)

summary(modelo_linear)
```

```r
plot(dados_ohe$Age, dados_ohe$Cholesterol,
     main = "Dispersão entre Idade e Colesterol",
     xlab = "Idade (normalizada)", ylab = "Colesterol (normalizado)",
     pch = 19, col = "darkgray")

abline(lm(Cholesterol ~ Age, data = dados_ohe),
       col = "blue", lwd = 2)
```

```r
par(mfrow = c(2,2))
plot(modelo_linear)
par(mfrow = c(1,1))
```

```r


coefs <- summary(modelo_linear)$coefficients
df_coef <- data.frame(
  Variavel = rownames(coefs),
  Coef = coefs[,1]
)

ggplot(df_coef[-1,], aes(x = reorder(Variavel, Coef), y = Coef)) +
  geom_bar(stat="identity") +
  coord_flip() +
  labs(title="Importância dos Coeficientes – Regressão Linear",
       x="", y="Coeficiente")
```

```r
set.seed(123)

treino_index <- createDataPartition(dados$HeartDisease, p = 0.8, list = FALSE)

treino <- dados[treino_index, ]
teste  <- dados[-treino_index, ]
```

```r
modelo_log <- glm(
  HeartDisease ~ .,
  data = treino,
  family = binomial(link = "logit")
)

summary(modelo_log)
```

```r
stepwise_aic <- function(X, y, verbose = TRUE) {
  dados <- data.frame(y = y, X)

  formula_nula <- as.formula("y ~ 1")
  formula_completa <- as.formula(paste("y ~", paste(names(X), collapse=" + ")))

  modelo_step <- step(
    glm(formula_nula, data = dados, family = binomial),
    scope = list(lower=formula_nula, upper=formula_completa),
    direction="both",
    trace = ifelse(verbose,1,0)
  )

  return(modelo_step)
}

X <- subset(treino, select = -HeartDisease)
y <- treino$HeartDisease

modelo_final <- stepwise_aic(X, y)
summary(modelo_final)
```

```r


prob <- predict(modelo_log, newdata = teste, type = "response")

roc_obj <- roc(teste$HeartDisease, prob)

plot(roc_obj, main = "Curva ROC – Regressão Logística")
auc(roc_obj)
```

```r
pred_class <- ifelse(prob > 0.5, 1, 0)

caret::confusionMatrix(
  factor(pred_class),
  factor(teste$HeartDisease),
  positive = "1"
)
```

```r

coefs <- summary(modelo_log)$coefficients
df <- data.frame(
  Variavel = rownames(coefs),
  OR = exp(coefs[,1])
)

ggplot(df[-1,], aes(x = reorder(Variavel, OR), y = OR)) +
  geom_bar(stat="identity") +
  coord_flip() +
  labs(title="Odds Ratio (Importância das Variáveis)",
       x="", y="Odds Ratio")
```

```r
plot(treino$Age, modelo_log$fitted.values,
     xlab="Idade (normalizada)", ylab="Probabilidade prevista",
     pch=19, col="darkblue",
     main="Probabilidade de Doença cardíaca por Idade")
```

```r
preprocess_norm <- preProcess(treino, method = c("center", "scale"))

treino_norm <- predict(preprocess_norm, treino)
teste_norm  <- predict(preprocess_norm, teste)
```

```r
modelo_knn <- train(
    HeartDisease ~ .,
    data = treino_norm,
    method = "knn",
    tuneLength = 10
)

modelo_knn
plot(modelo_knn)
```

```r
pred_knn <- predict(modelo_knn, teste_norm)
confusionMatrix(pred_knn, teste_norm$HeartDisease)
```

```r
modelo_arvore <- rpart(
    HeartDisease ~ .,
    data = treino,
    method = "class",
    parms = list(split = "gini")
)

rpart.plot(modelo_arvore, main = "Árvore de Decisão (Gini)")
```

```r
pred_arvore <- predict(modelo_arvore, teste, type = "class")
confusionMatrix(pred_arvore, teste$HeartDisease)
```

```r
modelo_rf <- randomForest(
    HeartDisease ~ .,
    data = treino,
    ntree = 500,
    mtry = 3,
    importance = TRUE
)

modelo_rf
varImpPlot(modelo_rf)
```
